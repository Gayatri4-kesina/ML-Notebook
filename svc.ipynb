```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
```

```python
Df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
Df.head()
```

```python
Df.info()
Df['species'].unique()
```

```python
Df.species.replace(('setosa','versicolor','virginica'),(1,2,3),inplace=True)
```

```python
plt.figure(figsize=(10,5))
Df.boxplot(color='red')
```

```python
Df1=Df[Df.species==1]
Df2=Df[Df.species==2]
Df3=Df[Df.species==3]
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(x=Df1['sepal_length'],y=Df1['sepal_width'], color='green')
plt.scatter(x=Df2['sepal_length'],y=Df2['sepal_width'], color='blue')
plt.scatter(x=Df3['sepal_length'],y=Df3['sepal_width'], color='red')
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(x=Df1['petal_length'],y=Df1['petal_width'], color='green')
plt.scatter(x=Df2['petal_length'],y=Df2['petal_width'], color='blue')
plt.scatter(x=Df3['petal_length'],y=Df3['petal_width'], color='red')
```

```python
x = Df.drop('species',axis=1)
y = Df['species']
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=104,train_size=0.4,shuffle=True)
model=SVC()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)  # PCA for train data
x_test_pca = pca.transform(x_test)        # PCA for test data

# Scatter plot for training data
plt.scatter(x_train_pca[:, 0], x_train_pca[:, 1], c=y_train, cmap='Blues', label='Train', alpha=0.6)

# Scatter plot for testing data
plt.scatter(x_test_pca[:, 0], x_test_pca[:, 1], c=y_test, cmap='Reds', label='Test', alpha=0.6)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/refs/heads/master/dataset.csv")
df.head()
```

```python
df.info()
df["target"].unique()
```

```python
X = df.drop("target", axis=1)
y = df["target"]                 #splitting
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.3,random_state=42)    #train test split
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)              #scaling
```

```python
log_clf = LogisticRegression()
dt_clf = DecisionTreeClassifier()
svc_clf = SVC(probability=True)           #MODELS
```

```python
voting_clf = VotingClassifier(estimators=[
    ('log_reg', log_clf),
    ('dt', dt_clf),
    ('svc', svc_clf)
], voting='soft')                           #soft voting
```

```python
start_time = time.time()
voting_clf.fit(X_train_scaled , y_train)
```

```python
training_time = time.time() - start_time
start_time = time.time()
```

```python
y_pred = voting_clf.predict(X_test_scaled)
prediction_time = time.time() - start_time
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(accuracy)
print(precision)
print(recall)
print(f1)
print("Confusion Matrix:")
print(conf_matrix)
print(f"Time taken for training: {training_time:.4f} seconds")
print(f"Time taken for prediction: {prediction_time:.4f} seconds")
```

```python
voting_clf_hard = VotingClassifier(estimators= [ ('log_reg', log_clf),
    ('dt', dt_clf),
    ('svc', svc_clf)], voting='hard')                                    #HARD VOTING
```

```python
start_time = time.time()
voting_clf_hard.fit(X_train_scaled,y_train)
```

```python
training_time = time.time() - start_time
start_time = time.time()
```

```python
y_pred_hard = voting_clf_hard.predict(X_test_scaled)
prediction_time = time.time() - start_time
```

```python
accuracy = accuracy_score(y_test, y_pred_hard)
precision = precision_score(y_test, y_pred_hard)
recall = recall_score(y_test,  y_pred_hard)
f1 = f1_score(y_test,  y_pred_hard)
conf_matrix = confusion_matrix(y_test,  y_pred_hard)

print(accuracy)
print(precision)
print(recall)
print(f1)
print("Confusion Matrix:")
print(conf_matrix)

print(f"Time taken for training: {training_time:.4f} seconds")
print(f"Time taken for prediction: {prediction_time:.4f} seconds")
```

```python
base_learners = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('svc', SVC(probability=True, random_state=42)),
    ('log_reg', LogisticRegression(random_state=42))                #STACKING
]
meta_model = LogisticRegression(random_state=42) 
```

```python
stack_clf = StackingClassifier(estimators = base_learners, final_estimator = meta_model)
```

```python
start_time= time.time()
stack_clf.fit(X_train_scaled,y_train)
training_time = time.time()-start_time
```

```python
start_time = time.time()
y_pred_stack = stack_clf.predict(X_test_scaled)
prediction_time = time.time() - start_time
```

```python
accuracy = accuracy_score(y_test,y_pred_stack)
precision = precision_score(y_test,y_pred_stack)
f1 = f1_score(y_test,y_pred_stack)
conf_mat = confusion_matrix(y_test,y_pred_stack)

print(accuracy)
print(precision)
print(recall)
print(f1)
print("Confusion Matrix:")
print(conf_matrix)

print(f"Time taken for training: {training_time:.4f} seconds")
print(f"Time taken for prediction: {prediction_time:.4f} seconds")
```

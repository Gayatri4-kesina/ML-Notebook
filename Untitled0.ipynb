```python
import numpy as np
import os
import cv2
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import pandas as pd
import warnings
```

```python
video_path = 'video'
output_folder = 'frames'
os.makedirs(output_folder, exist_ok=True)
```

```python
for video_file in os.listdir(video_path):
    if video_file.endswith(".mp4") or video_file.endswith(".avi"):
        video_full_path = os.path.join(video_path, video_file)
        cap = cv2.VideoCapture(video_full_path)

        if not cap.isOpened():
            print(f" Unable to open {video_file}")
            continue


        frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = cap.get(cv2.CAP_PROP_FPS)
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        duration = frame_count / fps if fps else 0

        print(f" Video: {video_file}")
        print(f"   - Resolution: {frame_width}x{frame_height}")
        print(f"   - FPS: {fps}")
        print(f"   - Total Frames: {frame_count}")
        print(f"   - Duration: {duration:.2f} seconds\n")

        cap.release()
```

```python
for video_file in os.listdir(video_path):
    if video_file.endswith(".mp4") or video_file.endswith(".avi"):
        cap = cv2.VideoCapture(os.path.join(video_path, video_file))
        frame_count = 0

        if not cap.isOpened():
            print(f"Error opening {video_file}")
            continue



        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            if frame_count % 5 == 0:
                frame_filename = f"{video_file}_frame_{frame_count}.jpg"
                save_path = os.path.join(output_folder, frame_filename)
                cv2.imwrite(save_path, frame)


            frame_count += 1

        cap.release()

print(f"Frames extracted sucessfully")
```

```python
import shutil
import glob
import random
import os

# Define paths
frames_path = "/content/frames"
dataset_path = "/content/dataset"

# Get all image files
all_images = glob.glob(frames_path + "/*.jpg")  # Adjust if images are in PNG format

# Shuffle and split into train (80%) and validation (20%)
random.shuffle(all_images)
split_idx = int(0.8 * len(all_images))

train_images = all_images[:split_idx]
val_images = all_images[split_idx:]

# Copy images to respective folders
for img in train_images:
    shutil.copy(img, f"{dataset_path}/images/train/")

for img in val_images:
    shutil.copy(img, f"{dataset_path}/images/val/")

print(f"Copied {len(train_images)} images to train and {len(val_images)} images to val ")
```

```python
from ultralytics import YOLO

# Load pre-trained YOLOv8 model
model = YOLO("yolov8n.pt")


results = model.predict(source="/content/dataset/images/train", save=True, conf=0.3)
```

```python
import os
import numpy as np

# Ensure the labels directory exists
labels_dir = "/content/dataset/labels"
os.makedirs(labels_dir, exist_ok=True)  # âœ… This will create the directory if missing

# Function to convert bounding boxes to YOLO format
def convert_yolo_format(box, img_width, img_height):
    x_center = (box[0] + box[2]) / 2 / img_width
    y_center = (box[1] + box[3]) / 2 / img_height
    width = (box[2] - box[0]) / img_width
    height = (box[3] - box[1]) / img_height
    return f"0 {x_center} {y_center} {width} {height}"  # '0' is the default class ID

# Process each detected image
for r in results:
    img_path = r.path
    img_name = os.path.basename(img_path).split(".")[0]
    label_path = os.path.join(labels_dir, f"{img_name}.txt")

    img_width, img_height = r.orig_shape[1], r.orig_shape[0]

    with open(label_path, "w") as f:
        for box in r.boxes.xyxy:
            label = convert_yolo_format(box.cpu().numpy(), img_width, img_height)
            f.write(label + "\n")

print(" Labels generated successfully!")
```
